### 연령대별 인구 비율

변수 간 상관 관계가 매우 높으면 PCA는 상관 관계가 높은 변수를 결합하려고 시도하고 고차원 데이터에서 최대 분산 방향을 찾습니다. 분산을 최대로 보존할 수 있는 축을 찾는 것이 정보를 가장 적게 손실할 수 있다고 생각할 수 있습니다. 분산이 커야 데이터들 사이의 차이점이 명확해질테고, 그것이 우리의 모델을 더 좋은 방향으로 만들 수 있게 됩니다.

연령대별 인구 수로 하면 X1+X2+X3=X4 가 될 거 같아서 (설명변수 간 상관관계 발생) 소숫점으로 변환

```python
age_temp = age.iloc[:, 0:14]
age_temp.columns
```

```python
# 0~ 29 / 30~59 / 60~ 으로 분류
age_temp["0~29세"] = age_temp.iloc[:, 3:6].sum(axis=1)/age_temp["인구수"]
age_temp["30~59세"] = age_temp.iloc[:, 6:9].sum(axis=1)/age_temp["인구수"]
age_temp["60세 이상"] = age_temp.iloc[:, 9:].sum(axis=1)/age_temp["인구수"]
age_temp.head()
```

```python
income_age = pd.concat([income, age_temp["인구수"], age_temp.iloc[:, -3:]], axis=1)
income_age.head()
```

```python
final = pd.merge(market_name[['시장(상점가명)', '시군구']], income_age, on="시군구", how="left")
final.head()
```

```python
# 그래프용
```

```python
import matplotlib.pyplot as plt 
import numpy as np

explained_variance_ratio = pca.explained_variance_ratio_

def explained_variance_ratio_plot(explained_variance_ratio):
    x_axis = range(1, len(explained_variance_ratio)+1)
    plt.bar(x_axis, explained_variance_ratio,
            align = 'center', color='pink',label = 'Individual Explained Variance Ratio')
    plt.step(x_axis, np.cumsum(explained_variance_ratio),
             where = 'mid', color='red', label='Cumulative Explained Variance Ratio')
    plt.ylim(0, 1.1)
    plt.xticks(x_axis)
    plt.xlabel('Principal Components')
    plt.ylabel('Explained Variance Ratio')
    plt.grid()
    plt.show()

explained_variance_ratio_plot(explained_variance_ratio)
```

```python
# n_components = 1
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = StandardScaler().fit_transform(final.iloc[:, 2:])
pca = PCA(n_components=1).fit(X)
custom = pca.transform(X)
```
