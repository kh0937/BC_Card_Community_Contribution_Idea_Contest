# PCA

```python
final.isnull().sum()
```

```python
# n_components = 2
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = StandardScaler().fit_transform(final.iloc[:, 2:])
pca = PCA(n_components=2).fit(X)
custom = pca.transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=final.iloc[:, 2:].columns)
```

```python
import matplotlib.pyplot as plt 
import numpy as np

explained_variance_ratio = pca.explained_variance_ratio_

def explained_variance_ratio_plot(explained_variance_ratio):
    x_axis = range(1, len(explained_variance_ratio)+1)
    plt.bar(x_axis, explained_variance_ratio,
            align = 'center', color='pink',label = 'Individual Explained Variance Ratio')
    plt.step(x_axis, np.cumsum(explained_variance_ratio),
             where = 'mid', color='red', label='Cumulative Explained Variance Ratio')
    plt.ylim(0, 1.1)
    plt.xticks(x_axis)
    plt.xlabel('Principal Components')
    plt.ylabel('Explained Variance Ratio')
    plt.grid()
    plt.show()

explained_variance_ratio_plot(explained_variance_ratio)
```

```python
# n_components = 1
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = StandardScaler().fit_transform(final.iloc[:, 2:])
pca = PCA(n_components=1).fit(X)
custom = pca.transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
pd.DataFrame(pca.components_.T, columns=['PC1'], index=final.iloc[:, 2:].columns)
```
