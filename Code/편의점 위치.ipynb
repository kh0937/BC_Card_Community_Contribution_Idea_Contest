```python
# example : 서울 강남구 삼성동 20km 반경에서 편의점 매장 검색
# 참고1 : https://developers.kakao.com/docs/latest/ko/local/dev-guide#search-by-category
# 참고2 : https://blog.naver.com/PostView.naver?blogId=blueqnpfr1&logNo=222069986010&parentCategoryNo=&categoryNo=1&viewDate=&isShowPopularPosts=true&from=search

import requests
import json

url = "https://dapi.kakao.com/v2/local/search/keyword.json?y=37.514322572335935&x=127.06283102249932&radius=20000" # x 경도 y 위도

queryString = {'query' : '편의점'}
header = {'Authorization': 'KakaoAK a2fba27f3df9c58f334f05f14025e5c7'}

response = requests.get(url, headers=header, params=queryString)
tokens = response.json() # json 형태로 output

print(response)
print(tokens)
```
